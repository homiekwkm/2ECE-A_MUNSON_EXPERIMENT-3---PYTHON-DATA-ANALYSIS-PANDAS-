# Problem 2

#### > For problem 2, we will be making use of the previous dataframe, and there are 4 specific data exctractions we need to make. Firstly, we will call the "cars" dataframe for reference:

```python
cars
```

#### > Generally, for all the following questions, there are two methods we can employ in order to extract the required data. We may make use of the .loc or .iloc commands to achieve our goals. Both will be demonstrated for each question.

#### a. Display the first five rows with odd-numbered columns (columns 1, 3, 5, 7...) of cars.

##### We can first make use of the .loc command to have this data extraction achieved

```python
#.loc - m row index, column name
#'Model' is displayed for reference
cars.loc[:4,['Model','mpg','disp','drat','qsec','am','carb']]
```

##### Then, we can make use of the .iloc command

```python
#.iloc - m row, n column indexes
#'Model' is displayed for reference
cars.iloc[:5,[0,1,3,5,7,9,11]]
```

#### b. Display the row that contains the 'Model' of 'Mazda RX4'

##### .loc command

```python
#.loc - m row index, column name
cars.loc[[0],['Model']]
```

##### .iloc command

```python
#.iloc - m row, n column indexes
cars.iloc[[0],[0]]
```

#### c. How many cylinders (‘cyl’) does the car model ‘Camaro Z28’ have?

##### .loc command

```python
#.loc - m row index, column name
cars.loc[[23],['Model','cyl']]
```

##### .iloc command

```python
#.iloc - m row, n column indexes
cars.iloc[[23],[0,2]]
```

#### d. Determine how many cylinders ('cyl') and what gear type ('gear') do the car models 'Mazda RX4 Wag', 'Ford Pantera L' and 'Honda Civic' have.

##### .loc command

```python
#.loc - m row index, column name
#'Model' is displayed for reference
cars.loc[[1, 28, 18],['Model','cyl', 'gear']]
```

##### .iloc command

```python
#.iloc - m row, n column indexes
#'Model' is displayed for reference
cars.iloc[[1, 28, 18],[0,2,10]]
```
